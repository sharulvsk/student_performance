```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv(r"C:\From Destop\student_performance\data\student_performance.csv")

df.head()
```

```python

df.describe()

df.isnull().sum()
```

```python

plt.figure(figsize=(8, 5))
sns.histplot(df['MathGrade'], kde=True, bins=20, color='blue')
plt.title('Distribution of Math Grades')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python

features = ['StudyHours', 'Attendance', 'PastScores']
for feature in features:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df[feature], y=df['MathGrade'])
    plt.title(f'{feature} vs MathGrade')
    plt.xlabel(feature)
    plt.ylabel('MathGrade')
    plt.show()
```

```python

df[df.isnull().any(axis=1)]
df.fillna(df.mean(), inplace=True)
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='MathGrade')
plt.title('Outliers in Math Grades')
plt.show()
```
